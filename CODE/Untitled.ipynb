```python
i,j,codesused=1,2,[]
def codechoose(x):
    global i,j,codesused
    try:
        code=str(x[0]+x[i]+x[j])
        if code not in codesused:
            codesused.append(code)
            i,j=1,2
            return(code)
        else:
            j=j+1
            if(j<len(x)):
                codechoose(x)
            else:
                i=i+1
                j=i+1
                codechoose(x)
    except IndexError:
        codesused.append("123")
df["NEWCODE"]=df["newname"].apply(codechoose)       
```
