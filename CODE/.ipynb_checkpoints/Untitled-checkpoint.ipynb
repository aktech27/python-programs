```python
illegal=[]
l1=list(df["newname"])
for i in l1:
    for j in range(0,len(i)):
        if i[j] not in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" :
            illegal.append(i[j])
print(list(set(illegal)))            
```

```python
df["newname"]=df["newname"].apply(lambda x: x.replace("Ô","O"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ó","O"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ö","O"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Î","I"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ï","I"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Í","I"))
df["newname"]=df["newname"].apply(lambda x: x.replace("É","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ê","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("È","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ë","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ê","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ě","E"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Â","A"))
df["newname"]=df["newname"].apply(lambda x: x.replace("À","A"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Á","A"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ä","A"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Û","U"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ü","U"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ů","U"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ú","U"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ç","C"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Č","C"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ÿ","Y"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ý","Y"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Œ","OE"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ť","T"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ř","R"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ž","Z"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ď","D"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Ň","N"))
df["newname"]=df["newname"].apply(lambda x: x.replace("Š","S"))
df
```

```python
i,j,codesused=1,2,[]
def codechoose(x):
    global i,j,codesused
    try:
        code=str(x[0]+x[i]+x[j])
        if code not in codesused:
            codesused.append(code)
            i,j=1,2
            return(code)
        else:
            j=j+1
            if(j<len(x)):
                codechoose(x)
            else:
                i=i+1
                j=i+1
                codechoose(x)
    except IndexError:
        codesused.append("123")
df["NEWCODE"]=df["newname"].apply(codechoose)       
```
