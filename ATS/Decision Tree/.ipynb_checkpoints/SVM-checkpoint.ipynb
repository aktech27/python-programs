```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris=load_iris()
iris
```

```python
dir(iris)
```

```python
print(iris.feature_names)
print(iris.target)
print(iris.target_names)
print(iris.data)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df['name']=df.target.apply(lambda x:iris.target_names[x])
df
```

```python
indep=df.drop(['target','name'],axis="columns")
dep=df['target']
```

```python
sepo=df[df.target==0]
vers=df[df.target==1]
virg=df[df.target==2]
```

```python
plt.scatter(sepo['sepal length (cm)'],sepo['sepal width (cm)'],marker='.',color='r',label="setosa")
plt.scatter(vers['sepal length (cm)'],vers['sepal width (cm)'],marker='.',color='g',label="versacolor")
plt.scatter(virg['sepal length (cm)'],virg['sepal width (cm)'],marker='.',color='b',label="virginica")
plt.legend()
plt.show()
```

```python
plt.scatter(sepo['petal length (cm)'],sepo['petal width (cm)'],marker='.',color='r',label="setosa")
plt.scatter(vers['petal length (cm)'],vers['petal width (cm)'],marker='.',color='g',label="versacolor")
plt.scatter(virg['petal length (cm)'],virg['petal width (cm)'],marker='.',color='b',label="virginica")
plt.legend()
plt.show()
```

```python
xtr,xts,ytr,yts=train_test_split(indep,dep,train_size=0.8)
```

```python
model=SVC()
model.fit(xtr,ytr)
model.score(xtr,ytr)
```

```python
model.predict(xts)
```
