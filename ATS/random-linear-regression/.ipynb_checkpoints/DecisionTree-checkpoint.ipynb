```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv(r'salary.csv')
df
```

```python
indep=df.drop('salary_more_100k',axis='columns')
dep=df['salary_more_100k']
```

```python
indep
```

```python
dep
```

```python
en_comp=LabelEncoder()
en_job=LabelEncoder()
en_deg=LabelEncoder()
```

```python
indep['new_comp']=en_comp.fit_transform(indep['company'])
indep['new_job']=en_job.fit_transform(indep['job'])
indep['new_deg']=en_deg.fit_transform(indep['degree'])
```

```python
indep
```

```python
indep=indep.drop(['company','job','degree'],axis='columns')
```

```python
indep
```

```python
model=DecisionTreeClassifier()
```

```python
Xtr,Xts,Ytr,Yts=train_test_split(indep,dep,test_size=0.3)
```

```python
print(len(Xtr))
print(len(Xts))
```

```python
model.fit(Xtr,Ytr)
model.predict(Xts)
```

```python
Yts
```

```python
model.score(Xts,Yts)
```
