```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('CarPrice.csv')
df1
```

```python
df1.columns
```

```python
reg=LinearRegression()
```

```python
reg.fit(df1[['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']],df1.price)
```

```python
independent=['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']
```

```python
plt.plot(df1[independent],reg.predict(df1[independent]))
plt.show()
```
