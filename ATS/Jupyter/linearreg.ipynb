```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df1=pd.read_csv('Book1.csv')
df2=pd.read_csv('Book2.csv')
df3=pd.read_csv('Book3.csv')
```

```python
reg1=linear_model.LinearRegression()
reg2=linear_model.LinearRegression()
reg1.fit(df1[['axis']],df1.value)
reg2.fit(df2[['axis']],df2.value)
```

```python
plt.plot(df1.axis,df1.value,color='y')
plt.scatter(df1.axis,df1.value,color='r',marker='o')
plt.show()
```

```python
plt.plot(df1.axis,df1.value,color='y',label='Plotting')
plt.plot(df1.axis,reg1.predict(df1[['axis']]),color='k',label="L.Reg")
plt.scatter(df1.axis,df1.value,color='r',marker='o')
plt.xlabel('AXIS')
plt.ylabel('VALUE')
plt.legend()
plt.show()
```

```python
plt.plot(df2.axis,df2.value,color='y',label='Plotting')
plt.plot(df2.axis,reg2.predict(df2[['axis']]),color='k',label="L.Reg")
plt.scatter(df2.axis,df2.value,color='r',marker='.')
plt.xlabel('AXIS')
plt.ylabel('VALUE')
plt.legend()
plt.show()
```
